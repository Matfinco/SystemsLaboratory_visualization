# Non linear ODE y' = -(y³ +2) solutions for different initial conditions

**Description:**

Left Plot: We can see on the left the plot of the solution, y(t), of a non liner ( look at the "+2" ) Ordinary Differential Equation (ODE) with different initial condition, for each of them we have a different euation that describes y(t).
We can see how all trajectories converge to (around) y = -1.25, which is the y_eq value for which y' = f(y_eq) = 0

Rigth Plot: On the rigth there is the plot of the actual non linear ODE described from the equation y' = -(y³+2).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definizione dell'ODE non lineare: y' = f(y, u)
def f(y, t):
    return -(y**3 + 2)  # Esempio di ODE non lineare

# Intervallo di tempo
t = np.linspace(0, 10, 1000)
y = np.linspace(-2, 2, 10000)

# Diversi valori iniziali y0
# y0_values = np.linspace(-2, 1, 10)  # 10 valori iniziali tra -1.5 e 1.5
y0_values = [-3, -2, 1, 1.5, 4]

# Creazione della figura con due sottografi
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# --- Grafico delle soluzioni nel tempo ---
for y0 in y0_values:
    sol = odeint(f, y0, t)
    axs[0].plot(t, sol, label=f'Y0 = {y0:.2f}')

axs[0].set_title("Non linear ODE y' = -(y³ +2) solutions for different initial conditions ")
axs[0].set_xlabel("t")
axs[0].set_ylabel("y(t)")
axs[0].legend(loc='best', fontsize='small')
axs[0].grid(True)

# --- Ritratto di fase (y vs y') ---
for y0 in y0_values:
    axs[1].plot(y, -(y**3 + 2), label=f"y'=f(y)")

axs[1].set_title("ODE function")
axs[1].set_xlabel("y")
axs[1].set_ylabel("y'")
axs[1].grid(True)

# Mostra il grafico
plt.tight_layout()
plt.show()
```
